```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.anova import anova_lm
df = pd.read_csv("E:\DATA SCIENCE MAJOR\CODES\TeachingRatings(p7).csv")
df
```

```python
#PROBLEM 1
```

```python
# =============================================================================
# Q1: Regression with T-test - Does gender affect teaching evaluation rates?
# =============================================================================

print("Q1. REGRESSION WITH T-TEST: Does gender affect teaching evaluation rates?")
print("="*70)

df['female'] = (df['gender'] == 'female').astype(int)

model1 = ols('eval ~ female', data=df).fit()
print("Regression Results:")
print(model1.summary())

# Additional statistical verification
print("\nAdditional Verification:")
male_scores = df[df['female'] == 0]['eval']
female_scores = df[df['female'] == 1]['eval']

t_stat, p_val = stats.ttest_ind(male_scores, female_scores)
print(f"Independent t-test results:")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_val:.4f}")
print(f"Mean evaluation - Male: {male_scores.mean():.3f}, Female: {female_scores.mean():.3f}")

print("\nInterpretation:")
if p_val < 0.05:
    print("There is a statistically significant difference in evaluation scores between genders.")
else:
    print("No significant difference found in evaluation scores between genders.")
```

```python
# =============================================================================
# Q2: Regression with ANOVA - Does beauty score differ by age group?
# =============================================================================

print("\nQ2. REGRESSION WITH ANOVA: Does beauty score differ by age group?")
print("="*70)

# Create age groups
df['age_group'] = pd.cut(df['age'], bins=[20, 35, 50, 65], 
                        labels=['Young', 'Middle', 'Senior'])

print("First Approach: One-way ANOVA")
# One-way ANOVA
model2 = ols('beauty ~ age_group', data=df).fit()
anova_results = anova_lm(model2)
print("ANOVA Results:")
print(anova_results)

# Group statistics
print("\nBeauty Score Statistics by Age Group:")
age_group_stats = df.groupby('age_group')['beauty'].agg(['mean', 'std', 'count'])
print(age_group_stats)
```

```python
# =============================================================================
# Q3: Correlation - Is teaching evaluation score correlated with beauty score?
# =============================================================================

print("\nQ3. CORRELATION: Is teaching evaluation score correlated with beauty score?")
print("="*70)

print("First Approach: Correlation Analysis")
# Correlation analysis
correlation_coef = df['eval'].corr(df['beauty'])
print(f"Pearson Correlation Coefficient: {correlation_coef:.4f}")

# Test significance
corr_test = stats.pearsonr(df['eval'], df['beauty'])
print(f"Correlation test - R: {corr_test[0]:.4f}, P-value: {corr_test[1]:.6f}")

# Regression model
model3 = ols('eval ~ beauty', data=df).fit()
print("\nRegression Results:")
print(model3.summary())
```

```python
# =============================================================================
# Summary of Findings
# =============================================================================

print("\n" + "="*70)
print("SUMMARY OF KEY FINDINGS")
print("="*70)

print("1. GENDER AND EVALUATION SCORES:")
print(f"   - Female coefficient in regression: {model1.params['female']:.4f}")
print(f"   - P-value for gender effect: {model1.pvalues['female']:.4f}")
print(f"   - Conclusion: {'Significant' if model1.pvalues['female'] < 0.05 else 'Not significant'} effect")

print("\n2. AGE GROUP AND BEAUTY SCORES:")
print(f"   - F-statistic from ANOVA: {anova_results['F']['age_group']:.4f}")
print(f"   - P-value from ANOVA: {anova_results['PR(>F)']['age_group']:.6f}")
print(f"   - Conclusion: {'Significant' if anova_results['PR(>F)']['age_group'] < 0.05 else 'Not significant'} difference")

print("\n3. CORRELATION BETWEEN EVALUATION AND BEAUTY:")
print(f"   - Correlation coefficient: {correlation_coef:.4f}")
print(f"   - P-value for correlation: {corr_test[1]:.6f}")
print(f"   - Conclusion: {'Significant' if corr_test[1] < 0.05 else 'Not significant'} correlation")

print("\n" + "="*70)
print("ANALYSIS COMPLETED")
print("="*70)
```
